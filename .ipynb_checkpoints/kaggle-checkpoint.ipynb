```python
#importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
# reading the data
df = pd.read_csv("kaggle1.csv")# to run the code use full path of file 
df.head()
df_train = pd.read_csv("kaggle_train1.csv",)#to run the code use full path of file 
df_train.head()
```

```python
#checking for NaN values in data 
df.isna().sum()
df_train.fillna(df["y"].mean,inplace=True)
df_train.isna().sum()
# found in data_train file
```

```python
sns.scatterplot(data = df,x="x",y="y")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
model = LinearRegression()
X = df[["x"]]
y = df["y"]


model.fit(X,y)
```

```python
x_test = df_train[["x"]]
y_test = df_train["y"]

y_pred = model.predict(x_test)
```

```python
import numpy as np
pd.DataFrame([y_test,y_pred])

type(y_pred),type(y_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
r2 = r2_score(y_test,y_pred)
```
