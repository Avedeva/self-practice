```python
#import necessary libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

df_test = pd.read_csv("C:\\Users\\suees\\skill\\ML\\Core ML\\Supervised Learning\\linear regression\\self practice\\kaggle1.csv")

df_train = pd.read_csv("C:\\Users\\suees\\skill\\ML\\Core ML\\Supervised Learning\\linear regression\\self practice\\kaggle_train1.csv")
df_test.head(),df_train.head()
```

```python
#checking for NaN values
df_train.isna().sum(),df_test.isna().sum()
```

```python
df_train.dropna(inplace=True)
df_train.isna().sum()
# droping the NaN value
```

```python
df_train.shape # checking the shape of the training data
```

```python
# plotting the graph to visulaize the data
sns.scatterplot(data = df_test,x="x",y="y")
xlabel = "x"
ylabel = "y"
plt.grid()
plt.legend()
plt.show()
```

```python
# now initiallizing the variables to train the model using linearRegression
X_train = df_train[["x"]]
y_train = df_train["y"]
from sklearn.linear_model import LinearRegression# importing the LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#testing variables
X_test = df_test[["x"]]
#point to remember the X parameter must be a dataframe and y can be array or panda series  
y_test = df_test["y"]
```

```python
# testing the Model
X_test = df_test[["x"]]
y_test = df_test["y"]
y_predict = model.predict(X_test)
from sklearn.metrics import r2_score #Evaluating the model using R-squared
r2_score(y_test,y_predict)# 0.98 score being 1 the best almost 98 percent accurate
```

```python
# showing the best fit line for the data 
sns.scatterplot(data = df_test,x="x",y="y",label = "Actual")
xlabel = "x"
ylabel = "y"
plt.grid()
plt.plot(X_test,y_predict,color = "Red",label = "predicted")


plt.legend()
plt.show()
```
