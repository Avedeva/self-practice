```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df["number"].value_counts().plot(kind='bar')
```

```python
df["fam_mem"] = pd.to_numeric(df["number"],errors='coerce',downcast="integer")
df["Alone_aah"] = np.where(df["fam_mem"].isnull(),df["number"],"nigga got family")
```

```python
df
```

```python
df["boggie"] = df["Cabin"].str[0]
df["seat_num"] = df["Cabin"].str.extract(r'(\d+)')
```

```python
df
```

```python
df["num_ticket"] = df["Ticket"].apply(lambda s:s.split()[-1])
df["num_ticket"] = pd.to_numeric(df["num_ticket"],errors="coerce",downcast="integer")

df["cat_ticket"] = df["Ticket"].apply(lambda s:s.split()[0])
```
