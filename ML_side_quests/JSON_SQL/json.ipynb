```python
import requests
response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=d7b3d75e1ed3c8b014644e8bcee75a21&language=en-US&page=2")
```

```python
df = pd.DataFrame(response.json()['results'])
df
```

```python
df = df[["adult","popularity","original_language","release_date","title","vote_average","vote_count"]]
df.head()
df.dtypes
```

```python
dfl=[]
for i in range(1,500):
    response = requests.get(f"https://api.themoviedb.org/3/movie/popular?api_key=d7b3d75e1ed3c8b014644e8bcee75a21&language=en-US&page={i}")

    temp_df = pd.DataFrame(response.json()['results'])
    temp_df = temp_df[["adult","popularity","original_language","release_date","title","vote_average","vote_count"]]
    dfl.append(temp_df)
    
    

    
```

```python
df=pd.concat(dfl,ignore_index=True)
```

```python
df.to_csv("Popular_movies.csv")
```
