```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df = df[["Age","EstimatedSalary","Purchased"]]
```

```python
df.head()
y = df["Purchased"]
X = df[["Age","EstimatedSalary"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.3,random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
class std:
    def __init__(self):
            pass
    def fit(self,X_train):
        X_mean_age = np.mean(X_train["Age"])
        X_mean_salary = np.mean(X_train["EstimatedSalary"])
        X_sd_age = np.std(X_train["Age"])
        X_sd_salary = np.std(X_train["EstimatedSalary"])


        X_train = np.array(X_train)

        X_scaled = np.zeros(X_train.shape)

        for i in range(X_train.shape[0]):
            temp = (X_train[i][0] - X_mean_age)/X_sd_age
            X_scaled[i][0] = temp
            

                   
        for i in range(X_train.shape[0]):
            temp = (X_train[i][1] - X_mean_salary)/X_sd_salary
            X_scaled[i][1] = temp
            
        return X_scaled

```

```python
std  = std()
X_std_train = std.fit(X_train)
X_std_test = std.fit(X_test)

```

```python
X_std_train  = pd.DataFrame(X_std_train,columns=X_train.columns)
X_std_test  = pd.DataFrame(X_std_test,columns=X_train.columns)
```

```python
X_std_train.head()
```

```python
fig,(ax1,ax2)  = plt.subplots(ncols=2,figsize = (12,5))
print("The Train Data")
ax1.scatter(X_train["Age"],X_train["EstimatedSalary"])
ax1.set_title("Before Scalling")
ax2.scatter(X_std_train["Age"],X_std_train["EstimatedSalary"],color = "red")
ax2.set_title("After Scalling")
plt.show()
```

```python
fig,(bx1,bx2)  = plt.subplots(ncols=2,figsize = (12,5))
print("The Test Data")
bx1.scatter(X_test["Age"],X_test["EstimatedSalary"])
bx1.set_title("Before Scalling")
bx2.scatter(X_std_test["Age"],X_std_test["EstimatedSalary"],color = "red")
bx2.set_title("After Scalling")
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize = (12,5))

ax1.set_title("Before Scalling")
ax2.set_title("After Scalling")

sns.kdeplot(x=X_train["Age"],ax=ax1,label = "Age")
sns.kdeplot(x=X_train["EstimatedSalary"],ax=ax1,label = "Salary")


sns.kdeplot(x=X_std_train["Age"],ax=ax2,label="Age")
sns.kdeplot(x=X_std_train["EstimatedSalary"],ax = ax2,label = "Salary")

plt.legend()

plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize = (12,5))

ax1.set_title("before Standardization")
ax2.set_title("After Scalling")

sns.kdeplot(x =  X_train["Age"],ax=ax1)

sns.kdeplot(x =  X_std_train["Age"],ax=ax2)

plt.show()
```
